## Bokeh: Applying Themes

```python
from bokeh.io import curdoc
from bokeh.plotting import figure, show, output_notebook

# Configure bokeh to use output in jupyter lab
output_notebook()
```

```python
# Prep data
x = [1,2,3,4,5]
y = [2,3,5,5,4]
```

```python
# Apply theme
# themes example: dark_minimal, light_minimal, night_sky, contrast, caliber
curdoc().theme = "contrast"
```

```python
# Create plot 
p = figure(sizing_mode="stretch_width", max_width=500, outer_height=250)
```

```python
# Add renderer 
p.line(x, y)

# Show plot
show(p)
```
