### Seaborn Crash Course

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.columns
```

```python
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker",
    style="smoker",
    size="size"
);
```

```python
dots = sns.load_dataset('dots')
sns.relplot(
    data=dots, 
    kind = "line",
    x = "time",
    y = "firing_rate",
    col="align",
    hue="choice",
    size="coherence",
    style="choice",
    facet_kws = dict(sharex=False)
)
```

### Statistical Estimation and Error Bars

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.columns
```

```python
sns.relplot(
    data = fmri,
    kind="line",
    x="timepoint",
    y="signal",
    col="region",
    hue="event",
    style="event",
);
```

### Linear Estimation with Seaborn

```python
sns.lmplot(
    data=tips,
    x="total_bill",
    y="tip",
    col="time",
    hue="smoker"
);
```

### Informative Distributional Summaries

```python
sns.displot(
    data=tips,
    x="total_bill",
    col="time",
    kde=True
);
```

```python
sns.displot(
    data=tips,
    kind="ecdf",
    x="total_bill",
    col="time",
    hue="smoker",
    rug=True
);
```

### Categorical Data

```python
sns.catplot(
    data=tips,
    kind="swarm",
    x="day",
    y="total_bill",
    hue="smoker"
);
```

```python
sns.catplot(
    data=tips,
    kind="violin",
    x="day",
    y="total_bill",
    hue="smoker",
    split=True
);
```

```python
sns.catplot(
    data=tips,
    kind="bar",
    x="day",
    y="total_bill",
    hue="smoker"
);
```

### Composite view onto multivariate datasets

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.columns
```

```python
sns.jointplot(
    data=penguins,
    x="flipper_length_mm",
    y="bill_length_mm",
    hue="species"
);
```

```python
sns.pairplot(
    data=penguins,
    hue="species"
);
```
