## Bokeh Legend, Texts and Annotations

```python
from bokeh.plotting import figure, show, output_notebook

# Configure bokeh to use output in jupyter lab
output_notebook()
```

```python
# prepare some data
x = [1, 2, 3, 4, 5]
y1 = [5, 4, 8, 3, 6]
y2 = [2, 6, 4, 5, 1]
y3 = [1, 2, 4, 6, 1]
```

```python
p = figure(title="Legend Example")
```

```python
l = p.line(x, y1, legend_label="Line", line_color="red", line_width=2)
c = p.circle(x, y2, legend_label="Circle", fill_color="blue", fill_alpha=0.5, size=75)
```

```python
# display legend in top left corner
p.legend.location = "top_left"

# Add Legend Title
p.legend.title = "My Legend Title"

# Change legend appearance
p.legend.label_text_font = "times"
p.legend.label_text_font_style = "bold"
p.legend.label_text_color = "navy"

# Change border and background of the legend
p.legend.border_line_width = 3
p.legend.border_line_color = "navy"
p.legend.border_line_alpha = 0.9
p.legend.background_fill_color = "yellow"
p.legend.background_fill_alpha = 0.2

show(p)
```

```python
x = [1,2,3,4,5]
y = [6,5,4,8,3]

p = figure(title="Main Headline Example")

p.line(x, y, legend_label="Line", line_width=2)

# Change headline location to left
p.title_location = "left"

# Change the Headline Text 
p.title.text = "HERE IS THE CHANGED TEXT"

# Style
p.title.text_font_size = "30px"
p.title.align = "right"
p.title.background_fill_color = "red"
p.title.text_color = "navy"
show(p)
```

```python
from bokeh.models import BoxAnnotation
import random
```

```python
# Generate random data
x = list(range(0, 101))
y = random.sample(range(0, 1000), 101)
```

```python
# Create Plot
p = figure(title="Box annotation example")
```

```python
l = p.line(x, y, line_color="blue", line_width=2)

# Add Box Annotation
low_box = BoxAnnotation(top=200, fill_alpha=0.1, fill_color="red")
high_box = BoxAnnotation(bottom=800, fill_alpha=0.1, fill_color="red")
mid_box = BoxAnnotation(top=800, bottom=200, fill_alpha=0.1, fill_color="blue")

# Add boxes to figure
p.add_layout(low_box)
p.add_layout(high_box)
p.add_layout(mid_box)

show(p)
```
