## Bokeh Custom Axes

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import NumeralTickFormatter

output_notebook()
```

```python
# Prep data
x = [1,2,3,4,5]
y = [2,4,6,8,10]
```

```python
# Create plot
p = figure(
    title = "Custom Axes",
    sizing_mode = "stretch_width",
    max_width = 600,
    height = 350
)
```

```python
# Add circle renderer
p.circle(x, y, size=15)

# Change the x-axis 
p.xaxis.axis_label = "I'm the X Axis"
p.xaxis.axis_line_width = 4
p.xaxis.axis_line_color = "red"

# Change the y-axis
p.yaxis.axis_label = "I am the Y Axis"
p.yaxis.major_label_text_color = "orange"
p.yaxis.major_label_orientation = "vertical"

# Change both axes
p.axis.minor_tick_in = -2
p.axis.minor_tick_out = 5
```

```python
# Show
show(p)
```

```python
x = [1,2,3,4,5]
y = [2,3,4,5,5]
```

```python
p = figure(
    title = "Tick Formatter",
    sizing_mode = "stretch_width",
    max_width = 800,
    height = 250
)
```

```python
# format axes ticks
p.yaxis[0].formatter = NumeralTickFormatter(format="$0.00")

# glyph
p.circle(x, y, size=5)
p.line(x, y, color="red", line_width=1)
```

```python
show(p)
```
